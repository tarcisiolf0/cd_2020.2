Assim, nos permite encontrar qual o modelo mais adequado para um determinado problema e base de dados entre os diversos algoritmos como árvore de decisão, SVM, naive bayes, etc. Além disso, nos permite achar configuração de hiperparâmetros mais próximo do ideal para o modelo, no caso de uma MLP, podemos por exemplo encontrar o número de camadas escondidas e seus  neurônios, além das ativações, para uma determinada aplicação. ## 7 - Usando os dados a seguir e um classificador, a qual classe o texto “what do you mean” deveria ser classificado? 

```python
Image(filename = "data2.jpeg", width=1000, height=1000)
```

```python
columns = ['Texto', 'Classe']
data = [["They are novels", "stmt"], ["have you read this book", "question"], ["who is the author", "question"], ["what are the characters", "question"], ["This is how I bought the book", "stmt"], ["I like fictions", "stmt"], ["what is your favorite book", "question"], ["This is my book", "stmt"]]
df = pd.DataFrame(data = data, columns = columns)
df
```

```python
df["Classe"] = df["Classe"].replace(['stmt', 'question'], [0,1])
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# create the document-term matrix with count vectorizer
corpus = df['Texto'].values
cv = CountVectorizer()
X = cv.fit_transform(corpus).toarray()
cv_dt = pd.DataFrame(X, columns=cv.get_feature_names())
cv_dt
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
Y = df['Classe'].values.astype('int')
model = MultinomialNB()
model.fit(X, Y)
```

```python
x = ['what do you mean']
X_p = cv.transform(x).toarray()
X_p
```

```python
predict = model.predict(X_p)
if predict == 0:
  print("A previsão foi STMT")
else:
  print("A previsão foi QUESTION")
```
